```python
#Reading Video from Webcam and Displaying it in Grayscale
video=cv2.VideoCapture(0)

while(True):
    ret,frame=video.read()
    grey=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow("Gray Scale Feed",grey)
    cv2.waitKey(1)
    if cv2.waitKey(33)==27:
        break
cv2.destroyAllWindows()
video.release()
    
```

```python
#Drawing Rectangle,Circle and Line on White Image
img1=np.ones((400,400,3),dtype='uint8')
img1[:,:]=[255,255,255]
rectangle=cv2.rectangle(img1,(30,30),(250,250),(0,0,255),-1)
line=cv2.line(img1,(50,50),(300,300),(255,125,23),4)
circle=cv2.circle(img1,(230,260),(58),(125,125,125),8)
cv2.imshow("White Image",img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Drawing Polygons and Ellipse
img2=np.zeros((500,500,3),dtype='uint8')
center_point=(int(img2.shape[0]/2),int(img2.shape[1]/2))

ellipse=cv2.ellipse(img2,center_point,(100,150),0,0,360,(120,23,45),4)

pts = np.array([[25, 70], [25, 160],  
                [110, 200], [200, 160],  
                [200, 70], [110, 20]], 
               np.int32)
ploygon=cv2.polylines(img2,[pts],True,(43,78,127),6)
cv2.imshow("Ploygon and Ellipse",img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
