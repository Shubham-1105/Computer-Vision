```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
rectangle=np.zeros((300,300),dtype='uint8')
cv2.rectangle(rectangle,(25,25),(275,275),(275,275,275),-1)
plt.imshow(rectangle)
plt.show()
```

```python
circle = np.zeros((300, 300), dtype = "uint8")
cv2.circle(circle, (150, 150), 150, (255, 255, 255), -1)
plt.imshow(circle)
plt.show()
```

```python
#Bitwise XOR Operation
bit_xor=cv2.bitwise_xor(rectangle,circle)
plt.imshow(bit_xor)
plt.show()
```

```python
#Bitwise Not Operation
bit_not=cv2.bitwise_not(rectangle,circle)
plt.imshow(bit_not)
plt.show()
```

```python
img=cv2.imread('anime.png')
```

```python
(h,w)=img.shape[:2]
(qh,qw)=(int(h/4),int(w/4))
```

```python
m=np.float32([[1,0,qh],[0,1,qh]])
```

```python
#Translation Down and Right
translated=cv2.warpAffine(img,m,(w,h))
plt.imshow(translated)
plt.show()
```

```python
hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(hsv)
plt.show()
```

```python
red=hsv[:,:,1]
plt.imshow(red)
plt.show()
```

```python
lower_red=np.array([0,0,0])
high_red=np.array([0,0,255])
```

```python
msk=cv2.inRange(hsv,lower_red,high_red)
```

```python
masked_image=np.copy(hsv)
```

```python
#Filtering the red Color 
masked_image[msk==0]=[0,0,0]
plt.imshow(masked_image)
plt.show()
```
